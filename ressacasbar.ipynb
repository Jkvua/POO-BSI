```
Apresentar o seu projeto (Diagrama de Classes) e implementar o sistema Drinks do Ressaca's Bar,
usando o padrão Decorator
• Imagine que você tem um drink chamado caipira, que custa 20,0 reais, porém   quando o cliente vai
  montando a sua caipira ele vai aumentando o preço
• Quero uma caipira de saquê, abacaxi, kiwi e açúcar
• Quero uma caipira de vodka, morango e adoçante
• Quero uma caipirinha (que é um drink pronto patenteado que não pode ser alterado – cachaça,
limão, gelo e açúcar)
```



````python
from abc import ABC, abstractmethod

class Drinks(ABC):
  def __init__(self, descricao, preco):
    self.descricao = descricao
    self.preco = preco

  @abstractmethod
  def getDescricao(self) -> str:
      pass

  @abstractmethod
  def getPreco(self) -> float:
      pass

class Saque(Drinks):
  def __init__(self):
      super().__init__("Saquê", 11.60)

  def getDescricao(self) -> str:
      return self.descricao

  def getPreco(self) -> float:
      return self.preco

class Vodka(Drinks):
  def __init__(self):
      super().__init__("Vodka pura", 12.60)

  def getDescricao(self) -> str:
      return self.descricao

  def getPreco(self) -> float:
      return self.preco

class Cerveja(Drinks):
  def __init__(self):
      super().__init__("Cerveja geladinha", 8.70)

  def getDescricao(self) -> str:
      return self.descricao

  def getPreco(self) -> float:
      return self.preco

class Whisky(Drinks):
  def __init__(self):
      super().__init__("Whisky com gelo", 16.0)

  def getDescricao(self) -> str:
      return self.descricao

  def getPreco(self) -> float:
      return self.preco

class Caipirinha(Drinks):
  def __init__(self):
      super().__init__("Uma caipirinha deliciosa", 20.0)

  def getDescricao(self) -> str:
      return self.descricao

  def getPreco(self) -> float:
      return self.preco

class DrinksDecoracao(Drinks):
  def __init__(self, drinks: Drinks):
      self.drinks = drinks

  def getDescricao(self) -> str:
      return self.drinks.getDescricao()

  def getPreco(self) -> float:
      return self.drinks.getPreco()

class Kiwi(DrinksDecoracao):
  def getDescricao(self) -> str:
      return self.drinks.getDescricao() + " + Adicionando Kiwi"

  def getPreco(self) -> float:
      return self.drinks.getPreco() + 9.50

class Abacaxi(DrinksDecoracao):
  def getDescricao(self) -> str:
      return self.drinks.getDescricao() + " + Adicionando Abacaxi"

  def getPreco(self) -> float:
      return self.drinks.getPreco() + 8.50

class Morango(DrinksDecoracao):
  def getDescricao(self) -> str:
      return self.drinks.getDescricao() + " + Adicionando Morango"

  def getPreco(self) -> float:
      return self.drinks.getPreco() + 6.50

class Limao(DrinksDecoracao):
  def getDescricao(self) -> str:
      return self.drinks.getDescricao() + " + Adicionando Limão"

  def getPreco(self) -> float:
      return self.drinks.getPreco() + 3.50

class Acucar(DrinksDecoracao):
  def getDescricao(self) -> str:
      return self.drinks.getDescricao() + " + Adicionando Açucar"

  def getPreco(self) -> float:
      return self.drinks.getPreco() + 2.50
````

````python
saque = Saque()
saque_com_abacaxi = Abacaxi(saque)
saque_com_abacaxi_kiwi = Kiwi(saque_com_abacaxi)
saque_com_abacaxi_kiwi_acucar = Acucar(saque_com_abacaxi_kiwi)

print("\nQuero uma caipira de saquê, abacaxi, kiwi e açúcar!")
print(f"{saque_com_abacaxi_kiwi_acucar.getDescricao()} \nPreço: {saque_com_abacaxi_kiwi_acucar.getPreco()}")

# 2
print("\nQuero uma caipira de vodka, morango e adoçante!")
vodka = Vodka()
vodka_com_morango = Morango(vodka)
vodka_com_morango_adocante = Acucar(vodka_com_morango)
print(f"{vodka_com_morango_adocante.getDescricao()} (adoçante como açúcar) \nPreço: {vodka_com_morango_adocante.getPreco()}")

# 3
print("\nQuero uma caipirinha (que é um drink pronto patenteado que não pode ser alterado – cachaça, limão, gelo e açúcar)!")
caipirinha_pronta = Caipirinha()
print(f"{caipirinha_pronta.getDescricao()} \nPreço: {caipirinha_pronta.getPreco()}")
````
