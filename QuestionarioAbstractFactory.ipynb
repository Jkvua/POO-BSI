````python
from abc import ABC, abstractmethod

class Chair(ABC):
    @abstractmethod
    def get_descricao(self):
        pass

class Cabinet(ABC):
    @abstractmethod
    def get_descricao(self):
        pass

class DiningTable(ABC):
    @abstractmethod
    def get_descricao(self):
        pass

class ScandinavianChair(Chair):
    def get_descricao(self):
        return "Uma cadeira escandinava moderna e minimalista."

class ScandinavianCabinet(Cabinet):
    def get_descricao(self):
        return "Um armário escandinavo elegante e funcional."

class ScandinavianDiningTable(DiningTable):
    def get_descricao(self):
        return "Uma mesa de jantar escandinava com um design simples."

class ClassicChair(Chair):
    def get_descricao(self):
        return "Uma cadeira clássica ornamentada."

class ClassicCabinet(Cabinet):
    def get_descricao(self):
        return "Um armário clássico com detalhes trabalhados."

class ClassicDiningTable(DiningTable):
    def get_descricao(self):
        return "Uma mesa de jantar clássica com acabamentos luxuosos."

class ContemporaryChair(Chair):
    def get_descricao(self):
        return "Uma cadeira contemporânea de design inovador."

class ContemporaryCabinet(Cabinet):
    def get_descricao(self):
        return "Um armário contemporâneo com linhas modernas."

class ContemporaryDiningTable(DiningTable):
    def get_descricao(self):
        return "Uma mesa de jantar contemporânea elegante e prática."

class FurnitureFactory(ABC):
    @abstractmethod
    def createChair(self) -> Chair:
        pass

    @abstractmethod
    def createCabinet(self) -> Cabinet:
        pass

    @abstractmethod
    def createDiningTable(self) -> DiningTable:
        pass

class ScandinavianFactory(FurnitureFactory):
    def createChair(self) -> Chair:
        return ScandinavianChair()

    def createCabinet(self) -> Cabinet:
        return ScandinavianCabinet()

    def createDiningTable(self) -> DiningTable:
        return ScandinavianDiningTable()

class ClassicFactory(FurnitureFactory):
    def createChair(self) -> Chair:
        return ClassicChair()

    def createCabinet(self) -> Cabinet:
        return ClassicCabinet()

    def createDiningTable(self) -> DiningTable:
        return ClassicDiningTable()

class ContemporaryFactory(FurnitureFactory):
    def createChair(self) -> Chair:
        return ContemporaryChair()

    def createCabinet(self) -> Cabinet:
        return ContemporaryCabinet()

    def createDiningTable(self) -> DiningTable:
        return ContemporaryDiningTable()

#Cliente
def cliente_codigo(factory: FurnitureFactory):
    chair = factory.createChair()
    cabinet = factory.createCabinet()
    dining_table = factory.createDiningTable()

    print(chair.get_descricao())
    print(cabinet.get_descricao())
    print(dining_table.get_descricao())

print("Scandinavian Factory:")
cliente_codigo(ScandinavianFactory())

print("\nClassic Factory:")
cliente_codigo(ClassicFactory())

print("\nContemporary Factory:")
cliente_codigo(ContemporaryFactory())
````

````python
from abc import ABC, abstractmethod

# produtos abstratos
class MotorVehicle(ABC):
    @abstractmethod
    def ligar_iguinicao(self):
        pass

class ElectricVehicle(ABC):
    @abstractmethod
    def ligar_motor(self):
        pass

# produtos concretos
class FutureVehicleMotorcycle(MotorVehicle):
    def ligar_iguinicao(self):
        return "FutureVehicle Motorcycle está ligando"

class FutureVehicleElectricCar(ElectricVehicle):
    def ligar_motor(self):
        return "FutureVehicle Electric Car está ligando"

# produtos concretos
class NextGenMotorcycle(MotorVehicle):
    def ligar_iguinicao(self):
        return "NextGen Motorcycle está ligando"

class NextGenElectricCar(ElectricVehicle):
    def ligar_motor(self):
        return "NextGen Electric Car está ligando"

# fábrica abstrata
class Corporation(ABC):
    @abstractmethod
    def createMotorVehicle(self) -> MotorVehicle:
        pass

    @abstractmethod
    def createElectricCar(self) -> ElectricVehicle:
        pass

# fábrica concreta
class FutureVehicleCorporation(Corporation):
    def createMotorVehicle(self) -> MotorVehicle:
        return FutureVehicleMotorcycle()

    def createElectricCar(self) -> ElectricVehicle:
        return FutureVehicleElectricCar()

# fábrica concreta
class NextGenCorporation(Corporation):
    def createMotorVehicle(self) -> MotorVehicle:
        return NextGenMotorcycle()

    def createElectricCar(self) -> ElectricVehicle:
        return NextGenElectricCar()

#cliente
def cliente_codigo(corporation: Corporation):
    motor_vehicle = corporation.createMotorVehicle()
    electric_vehicle = corporation.createElectricCar()

    print(motor_vehicle.ligar_iguinicao())
    print(electric_vehicle.ligar_motor())

print("FutureVehicleCorporation:")
cliente_codigo(FutureVehicleCorporation())

print("\nNextGenCorporation:")
cliente_codigo(NextGenCorporation())
````

````python
from abc import ABC, abstractmethod

class Weapon(ABC):
    @abstractmethod
    def use(self) -> str:
        pass

class Armor(ABC):
    @abstractmethod
    def protecao(self) -> str:
        pass

    @abstractmethod
    def usefulFunctionWithWeapon(self, weapon: Weapon) -> str:
        pass

class Axe(Weapon):
    def use(self) -> str:
        return "Você brandiu um machado feroz!"

class BodyArmor(Armor):
    def protecao(self) -> str:
        return "A armadura protegeu contra o golpe!"

    def usefulFunctionWithWeapon(self, weapon: Weapon) -> str:
        return f"BodyArmor utilizada com {weapon.use()}"

class MageFireBall(Weapon):
    def use(self) -> str:
        return "Você lançou uma bola de fogo devastadora!"

class ClothArmor(Armor):
    def protecao(self) -> str:
        return "O manto mágico reduziu o dano!"

    def usefulFunctionWithWeapon(self, weapon: Weapon) -> str:
        return f"ClothArmor utilizada com {weapon.use()}"

class OrcWeapon(Weapon):
    def use(self) -> str:
        return "O Orc brandiu sua arma brutal!"

class OrcArmor(Armor):
    def protecao(self) -> str:
        return "A armadura do Orc absorveu o impacto!"

    def usefulFunctionWithWeapon(self, weapon: Weapon) -> str:
        return f"OrcArmor utilizada com {weapon.use()}"

class AbstractFactory(ABC):
    @abstractmethod
    def create_weapon(self) -> Weapon:
        pass

    @abstractmethod
    def create_armor(self) -> Armor:
        pass

class WarriorFactory(AbstractFactory):
    def create_weapon(self) -> Weapon:
        return Axe()

    def create_armor(self) -> Armor:
        return BodyArmor()

class MageFactory(AbstractFactory):
    def create_weapon(self) -> Weapon:
        return MageFireBall()

    def create_armor(self) -> Armor:
        return ClothArmor()

class OrcFactory(AbstractFactory):
    def create_weapon(self) -> Weapon:
        return OrcWeapon()

    def create_armor(self) -> Armor:
        return OrcArmor()

class Cliente:
    def __init__(self, factory: AbstractFactory):
        self.weapon = factory.create_weapon()
        self.armor = factory.create_armor()

    def battle(self):
        print(self.weapon.use())
        print(self.armor.protecao())
        print(self.armor.usefulFunctionWithWeapon(self.weapon))

print("Guerreiro:")
warrior = Cliente(WarriorFactory())
warrior.battle()

print("\nMago:")
mage = Cliente(MageFactory())
mage.battle()

print("\nOrc:")
orc = Cliente(OrcFactory())
orc.battle()
````
