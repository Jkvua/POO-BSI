1) Há a necessidade de criar um esquema que utilize quatro formas bastante conhecidas de ordenação:
bubble sort, insertion sort, merge sort e quick sort. Coloque-as no padrão Strategy e
```
escreva um cliente que alterna a estratégia de ordenação livremente. (Apenas desenhar o diagrama que representa o padrão Strategy).

```

````python
from ast import In
from abc import ABC, abstractmethod

class SortStrategy(ABC):
    @abstractmethod
    def alterar(self, arr):
        pass

class BubbleSort(SortStrategy):
    def alterar(self, arr):
      n = len(arr)
      for i in range(n):
          for j in range(0, n-i-1):
              if arr[j] > arr[j+1] :
                  arr[j], arr[j+1] = arr[j+1], arr[j]
      print(f"Usando o bubble sort {arr}")

class InsertionSort(SortStrategy):
    def alterar(self, arr):
      for i in range(1, len(arr)):
          key = arr[i]
          j = i - 1
          while j >= 0 and key < arr[j]:
              arr[j + 1] = arr[j]
              j -= 1
          arr[j + 1] = key
      print(f"Usando o insertion sort {arr}")

class MergeSort(SortStrategy):
    def alterar(self, arr):
      self.merge_sort(arr)
      print(f"usando o merge sort {arr}")

    def merge_sort(self, arr):
      if len(arr) > 1:
          mid = len(arr) // 2
          L = arr[:mid]
          R = arr[mid:]

          self.merge_sort(L)
          self.merge_sort(R)

          i = j = k = 0
          while i < len(L) and j < len(R):
              if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
              else:
                arr[k] = R[j]
                j += 1
              k += 1

          while i < len(L):
              arr[k] = L[i]
              i += 1
              k += 1

          while j < len(R):
              arr[k] = R[j]
              j += 1
              k += 1

class QuickSort(SortStrategy):
    def alterar(self, arr):
      self.quick_sort(arr, 0, len(arr) - 1)
      print(f"usando o quick sort {arr}")

    def quick_sort(self, arr, low, high):
      if low < high:
          pi = self.partition(arr, low, high)
          self.quick_sort(arr, low, pi - 1)
          self.quick_sort(arr, pi + 1, high)

    def partition(self, arr, low, high):
      pivot = arr[high]
      i = low - 1
      for j in range(low, high):
          if arr[j] <= pivot:
              i = i + 1
              arr[i], arr[j] = arr[j], arr[i]
      arr[i + 1], arr[high] = arr[high], arr[i + 1]
      return i + 1

class Cliente:
  def __init__(self, strategy: SortStrategy):
    self.sortStrategy = strategy

  def setStrategy(self, strategy: SortStrategy):
    self.sortStrategy = strategy

  def alterar(self, arr):
    self.sortStrategy.alterar(arr)

if __name__ == "__main__":
  arr = [64, 25, 12, 22, 11]

  cliente = Cliente(BubbleSort())
  cliente.alterar(arr.copy())

  cliente.setStrategy(InsertionSort())
  cliente.alterar(arr.copy())

  cliente.setStrategy(MergeSort())
  cliente.alterar(arr.copy())

  cliente.setStrategy(QuickSort())
  cliente.alterar(arr.copy())
````

2) Considerando as classes abaixo, execute as seguintes atividades:
```
• Organize as hierarquias de classes adequadas
• Identifique uma classe abstrata, uma interface e oito classes
• Desenhe as setas dos relacionamentos entre as classes
• É_UM (IS-A)
• TEM_UM (HAS-A)
• IMPLEMENTA (Implements)
• Crie um método chamado atribuiArma() e coloque-o na classe certa
• O esquema de classes deve representar o padrão Strategy.
```

````python
from abc import ABC, abstractmethod

class Armas(ABC):
    @abstractmethod
    def usarArma(self, usar_arma):
      pass

class Espada(Armas):
  def usarArma(self):
    print(f"Usando a Espada")

class Faca(Armas):
  def usarArma(self):
    print(f"Usando uma Faca")

class ArcoeFlexa(Armas):
  def usarArma(self):
    print(f"Usando o Arco e Flexa")

class Machado(Armas):
  def usarArma(self):
    print(f"Usando o Machado")

class Personagem:
  def __init__(self):
    self.arma = None

  @abstractmethod
  def lutar(self):
    pass

  def atribuirArma(self, arma: Armas):
    self.arma = arma

class Rainha(Personagem):
  def lutar(self):
    print("Rainha está pronta para lutar")
    if self.arma:
      self.arma.usarArma()
    else:
      print("Rainha não tem arma")

class Duende(Personagem):
  def lutar(self):
    print("Duende está pronta para lutar")
    if self.arma:
      self.arma.usarArma()
    else:
      print("Duende não tem arma")

class Rei(Personagem):
  def lutar(self):
    print("Rei está pronta para lutar")
    if self.arma:
      self.arma.usarArma()
    else:
      print("Rei não tem arma")

class Guerreiro(Personagem):
  def lutar(self):
    print("Guerreiro está pronta para lutar")
    if self.arma:
      self.arma.usarArma()
    else:
      print("Guerreiro não tem arma")

if __name__ == "__main__":
  rainha = Rainha()
  duende = Duende()
  rei = Rei()
  guerreiro = Guerreiro()

  espada = Espada()
  faca = Faca()
  arco_flecha = ArcoeFlexa()
  machado = Machado()

  rainha.atribuirArma(espada)
  rainha.lutar()

  duende.atribuirArma(faca)
  duende.lutar()

  rei.atribuirArma(arco_flecha)
  rei.lutar()

  guerreiro.atribuirArma(machado)
  guerreiro.lutar()
````
